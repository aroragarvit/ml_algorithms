```python

  # samples in a rows and features in columns of dataset  
import numpy as np
import pandas as pd 

def normalize(df):
    result = df.copy()
    for column in df.columns:
        max_value = df[column].max()
        min_value = df[column].min()
        if(max_value-min_value!=0):
            result[column] = (df[column] - min_value) / (max_value - min_value) 
        else:
            result[column]=df[column]
    return(result)    
        
data=pd.read_csv("houseprice.csv")
data.columns = (["housesize", "rooms", "price"])

normalised_data=normalize(data)


result = normalised_data.to_csv("result.csv",columns=['price'],index=False)

train = normalised_data.to_csv("train.csv",columns=['housesize',"rooms"],index=False)



```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)

train=(pd.read_csv('train.csv'))                                 # norrmalised
train.insert(0,'ones',1)   # important function of pandas   

train=train.to_numpy()

result=(pd.read_csv('result.csv')).to_numpy()


epochs=[]
error=[]
class linear_regression:

     def __init__(self,train,result,epoches,alpha):
         self.train=train    #//  m*n+1
         self.result=result  #//  m*1
         samples,features=train.shape             # features will also contain 1-1-1-1 in them as first  column 
         self.samples=samples
         self.features=features
         self.weights=np.random.rand(features,1)  # n+1*1 # it will contain a single column and number of rows same as number of features
         self.epoches=epoches
         self.alpha=alpha
     def training(self):
         
         for i in range(self.epoches):
            epochs.append(i)
            prediction_matrix=np.dot(self.train,self.weights)  #// m*1                                                                                             # our predicted y result  according to our thetas  for features  of different samples
            mean_error=1/self.samples*(np.sum(np.square(np.subtract(prediction_matrix,self.result)),axis=0)) #// 1/2m*(sum of columns of square of elements of m*1-m*1 )      # column matrix with square error for different samples in each each row will do the sum of each of them to find overall error 
            print(" epochee${}-----------------Mean_square_error{}".format(i,mean_error))
            error.append(mean_error)
            weight_subtract_matrix =np.dot(np.transpose( np.subtract(prediction_matrix,self.result)),self.train)  #// (m*1 - m*1 )^T . m*n+1 = 1*n+1
            self.weights=np.subtract(self.weights,np.transpose(weight_subtract_matrix)*self.alpha/self.samples)  #// n+1*1 - n+1*1
        
a=linear_regression(train,result,epoches=250,alpha=0.01)

print(" random weights before training are")
print("-----------------------------------------------------------------------------------")
print(a.weights)
print("-----------------------------------------------------------------------------------")

a.training()

print("----------------------your final thetas after optimisation are-------------------------------------")
print(a.weights)

plt.title("Error With Each Epoches",loc = 'left')
plt.xlabel("epoches")
plt.ylabel("mean square error")
plt.plot(epochs,error)
plt.show()
```
